```python
df["Year"]=df["Date_fixed"].dt.year
```

```python
df["Month"]=df["Date_fixed"].dt.month
```

```python
df["Day"]=df["Date_fixed"].dt.day
```

```python
df["Monthname"]=df["Date_fixed"].dt.month_name()
```

```python
df["Dayname"]=df["Date_fixed"].dt.day_name()
```
