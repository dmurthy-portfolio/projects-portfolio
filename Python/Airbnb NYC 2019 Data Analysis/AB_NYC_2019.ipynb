```python
#Fill the missing null values of each column by the median value, except for the 'reviews_per_month' and last_review column.
df.fillna({'name':df['name'].mode()[0]}, inplace=True)
```

```python
df.fillna({'host_name':df['host_name'].mode()[0]},inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#Use mean method to replace the null values in "reviews_per_month"
df.fillna({'reviews_per_month':df['reviews_per_month'].mean()},inplace=True)
```

```python
df.isnull().sum()
```

```python
##explore ffill() and bfill() methods
df[df['last_review'].isnull()]
```

```python
df.index = range(1,48896)
```

```python
df
```

```python
##Sort the dataset by the 'price' column in descending order.
df.sort_values(["price"],ascending=False)
```

```python
df.sort_values(by = ['neighbourhood','price'])
```

```python
#Select and display the rows where the 'price' is greater than $500 for woodside
df[(df['price'] > 500) & (df['neighbourhood'] == 'woodside')]
```

```python
#select and display all listings from vailable Kensington and Midtown
df[(df['neighbourhood'] == 'Kensington') | (df['neighbourhood'] == 'Midtown')]
```

```python
df[df['neighbourhood'].isin(['Kensington', 'Midtown'])]
```

```python
## Calculate the price per night for each listing and create a new column 'price_per_night' in the DataFrame.
df['price_per_night'] = df['price']/df['minimum_nights']
```

```python
df.head()
```

```python
## convert the last_review to datetime val
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df
```

```python
df.info()
```

```python
df['last_review'] = df['last_review'].ffill()
```

```python
df.isnull().sum()
```

```python
#Group the data by the 'neighbourhood_group' column and calculate the mean price for each group. identify the high price
df.groupby('neighbourhood_group')[['price','price_per_night']].mean()
```

```python
df.groupby(['neighbourhood_group','neighbourhood'])[['price','price_per_night']].mean()
```

```python
#calculate min,avg,max,mean for the price of each neighbourhood_group
df.groupby('neighbourhood_group')['price'].agg(['min','max','mean','count'])
```

```python
#Count the number of listings in each neighbourhood group.
df['neighbourhood_group'].value_counts()
```

```python
df.groupby('neighbourhood_group').size().sort_values(ascending=False)
```

```python
## creat pivot_table that displays number of listing fr each neighbourhood_group for diff room types
df.pivot_table(index='neighbourhood_group',columns='room_type',values='id',aggfunc='count')
```

```python
## fetch the avg prices for each neighbourhood grp and for each room type
df.pivot_table(index='room_type',columns='neighbourhood_group',values='price',aggfunc='mean')
```
